```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import json
import ast
import psycopg2
import io
from sqlalchemy import create_engine
```

```python

# Request to website and download HTML contents
pagenumber=4
#Empty Dataframe
lazadapage_list = []
for i in range(1,pagenumber):
#put the url here, take note that it has to have the "page=" string  with "{}" after so that you can choose how many pages the program would scrape the data from.
    url='https://www.lazada.com.ph/catalog/?_keyori=ss&from=input&page={}&q=earphones&spm=a2o4l.home.search.go.45d8359dEg2I87'.format(i)
    req=requests.get(url)
    content=req.text
    soup=BeautifulSoup(content)
#Get JSON format
    raw=soup.find_all('script')[4].text.strip()[18:-381]
#Transform json to Python Dict format to access the data
    data=json.loads(raw)
    x = data["mods"]
    items = x["listItems"]
    table=pd.DataFrame.from_dict(items)
    lazadapage_list.append(table)
df = pd.concat(lazadapage_list,axis=0, ignore_index=True)

```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
display(df)
```
